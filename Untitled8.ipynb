```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Input, Dense, SimpleRNN
from tensorflow.keras.preprocessing.text import Tokenizer
import re
import numpy as np
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.ukrlib.com.ua/books/printit.php?tid=774#google_vignette"
request = requests.get(url)
html = BeautifulSoup(request.content, "html.parser")
html_texts = html.find("article", class_="prose")
texts = []
[texts.append(text.text) for text in html_texts]
```

```python
data_text = ""
for s in texts:
  data_text += s
```

```python
def prepare_data():
  with open("data.txt", "r", encoding="utf-8") as f:
    text = f.read()
    text = re.sub(r'[^А-ї]', ' ', text)
    return {"text":text, "vocab":sorted(set(text))}
```

```python
text = prepare_data()["text"]
print(text)
```

```python
symbol_amount = 34
tokenizer = Tokenizer(num_words=symbol_amount, char_level=True)
tokenizer.fit_on_texts([text])
print(tokenizer.word_index)

inp_chars = 6
data = tokenizer.texts_to_matrix(text)
n = data.shape[0] - inp_chars

print(n)
```

```python
X = np.array([data[i:i + inp_chars, :] for i in range(n)])
Y = data[inp_chars:]

print(data.shape)
```

```python
model = keras.models.Sequential([
    Input((
        inp_chars,
        symbol_amount
    )),
    SimpleRNN(256),
    Dense(symbol_amount, activation="softmax")
])
model.summary()
model.compile(loss="categorical_crossentropy",
          metrics=["accuracy"], optimizer="adam")
```

```python
history = model.fit(X, Y, batch_size=32, epochs=200)
```

```python
def buildPhrase(inp_str, str_len=150):
    for i in range(str_len):
        x = []
        for j in range(i, i + inp_chars):
            x.append(tokenizer.texts_to_matrix(inp_str[j]))

        x = np.array(x)
        inp = x.reshape(1, inp_chars, symbol_amount)

        pred = model.predict(inp)
        d = tokenizer.index_word[pred.argmax(axis=1)[0]]

        inp_str += d

    return inp_str
```

```python
res = buildPhrase("україн")
print(res)
```
